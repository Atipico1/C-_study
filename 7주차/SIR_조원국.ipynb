```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns  

import os

filename = "sir/0.800000"
file = np.loadtxt(filename+".txt")
epochs = len(file) // 40

print(epochs)

sarr = []
iarr = []
rarr = []

for ep in range(epochs):
    x = file[ep*40:(ep+1)*40]
    sarr.append((x == 0).mean())
    iarr.append((x == 1).mean())
    rarr.append((x == 2).mean())

    if ep % 20 == 0:
        print(ep)
        x = file[ep*40:(ep+1)*40]
        plt.figure()
        sns.heatmap(x, vmin=0, vmax=2)
        plt.show()

plt.figure()
plt.plot(sarr, color= 'darkblue', linewidth = 2, label = "S")
plt.plot(iarr, color= 'crimson', linewidth = 2, label = "I")
plt.plot(rarr, color = 'forestgreen', linewidth = 2, label = "R")
plt.legend()
plt.show()
```
